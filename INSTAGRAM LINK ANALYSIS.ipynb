```python
# prompt: instagram network with names and graphs

# Install necessary libraries
!pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

# Sample Instagram user data (replace with your actual data)
users = {
    "Alice": ["Bob", "Charlie", "David"],
    "Bob": ["Alice", "Eve", "Mallory"],
    "Charlie": ["Alice", "Trent"],
    "David": ["Alice", "Mallory"],
    "Eve": ["Bob"],
    "Mallory": ["Bob", "David"],
    "Trent": ["Charlie"]
}


# Create a graph
graph = nx.Graph()

# Add nodes (users)
for user in users:
  graph.add_node(user)

# Add edges (connections)
for user, connections in users.items():
  for connection in connections:
    graph.add_edge(user, connection)


# Visualize the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(graph) # positions for all nodes
nx.draw(graph, pos, with_labels=True, node_color="skyblue", node_size=1000, font_size=10, font_weight="bold", edge_color="gray", width=1.5)  # draw the graph with labels
plt.title("Instagram Network")
plt.show()

# Basic analysis (example: degree centrality)
degree_centrality = nx.degree_centrality(graph)
print("Degree Centrality:")
for user, centrality in degree_centrality.items():
  print(f"{user}: {centrality}")

#You can do other graph analysis, like:
nx.betweenness_centrality(graph) # betweenness centrality
nx.closeness_centrality(graph) # closeness centrality
nx.eigenvector_centrality(graph) # eigenvector centrality
```
