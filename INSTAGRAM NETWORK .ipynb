```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Generate a fake dataset of user interactions
def generate_fake_interactions(num_users, num_interactions):
    interactions = []
    for _ in range(num_interactions):
        user1 = random.randint(1, num_users)
        user2 = random.randint(1, num_users)
        if user1 != user2:
            interactions.append((f'user{user1}', f'user{user2}'))
    return interactions

# Parameters
num_users = 30
num_interactions = 60

# Generate fake interactions
fake_interactions = generate_fake_interactions(num_users, num_interactions)

# Create a graph
G = nx.Graph()

# Add edges to the graph based on interactions
for user1, user2 in fake_interactions:
    G.add_edge(user1, user2)



# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray', font_size=10, font_weight='bold')
plt.title('Instagram User Interaction Network')
plt.show()

```
